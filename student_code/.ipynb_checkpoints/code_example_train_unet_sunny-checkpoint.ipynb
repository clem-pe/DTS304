```python
# example unet model training code for sunny bright image set

# import related libraries
from dataset import camvidLoader
import numpy as np
from matplotlib import pyplot as plt
from dataset import from_label_to_rgb
import data_aug as aug
from skimage.io import imsave
import os
from unet import UNet
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
# hyper-parameters 
lr = 0.0002
epochs = 50
batch_size = 8
num_workers = 8
img_size = [384, 384]
# we are training on the sunny camvid dataset. 
data_root = '../CamVid/sunny'
num_classes = 14 # this is fixed for our project, we have 14 semantic classes
```

```python
# we first double confirm the dataloader works correctly
# we disable data argumentation and pytorch transformation to check the dataloader correctness
img_label_data = camvidLoader(root=data_root, split='train', is_aug=False, img_size = img_size, is_pytorch_transform=False)
img, label, img_path = img_label_data[0]
# this is to convert from the loaded label map to rgb label map visualization
label_rgb_vis = from_label_to_rgb(label) 
plt.figure(0) 
# original image visualization
plt.imshow(img) 
plt.figure(1)
# rgb label map visualization
# note that we have performed certain label conversion, so the color map not exactly matches original segmenation groundtruth provided by CamVid
plt.imshow(label_rgb_vis) 
```

```python
# during the training, we perform data argumentation. 
# This is to check the data argumented image is also reasonable.
aug_obj = aug.Compose([aug.RandomHorizontalFlip(), aug.RandomResizedCrop(img_size[0]),
                   aug.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4)])
img_label_data_aug = camvidLoader(root=data_root, split='train', is_aug=True, img_size = img_size, 
                                  is_pytorch_transform = False, aug = aug_obj)
for i in range(5):
    img, label, img_path = img_label_data_aug[i]
    label_rgb_vis = from_label_to_rgb(label) # this is to convert from the loaded label map to rgb label map visualization
    plt.figure(i * 2 + 0) 
    plt.imshow(img) 
    plt.figure(i * 2 + 1)
    plt.imshow(label_rgb_vis)
```

```python
# we set is_pytorch_transform to true, in order to perform network training/testing (see code for details)
# we also do data argumentation for training
train_dataset = camvidLoader(root=data_root, split='train', is_aug=True, img_size = img_size, 
                             is_pytorch_transform = True, aug = aug_obj) 
train_loader = DataLoader(train_dataset, num_workers=num_workers, batch_size=batch_size, shuffle=True, drop_last=True)
# we do not do data argumentation for validation
val_dataset = camvidLoader(root=data_root, split='val', is_aug=False, img_size = img_size, 
                             is_pytorch_transform = True, aug = None) 
val_loader = DataLoader(val_dataset, num_workers=num_workers, batch_size=batch_size, shuffle=False, drop_last=False)
```

```python
# Define our U-Net model, loss function and optimizer
device = 'cuda' # can be either set to cuda or cpu
unet = UNet(3, num_classes, width=32, bilinear=True)
unet = unet.to(device)
loss_func = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(unet.parameters(), lr=lr)
```

```python
# make the result directory
train_result_dir = '../train_internal_data/'
if not os.path.exists(train_result_dir):
    os.makedirs(train_result_dir)

# finally!!! the training loop!!!
for epoch in range(epochs):
    # training loop
    unet.train()
    train_loss = 0
    count = 0
    for idx_batch, (imagergb, labelmask, filename) in enumerate(train_loader):
        # zero the grad of the network before feed-forward
        optimizer.zero_grad()

        # send to the device (GPU or CPU) and do a forward pass
        x = imagergb.to(device)
        y_ = labelmask.to(device)
        y = unet(x)

        # finally calculate the loss and back propagate
        loss = loss_func(y, y_)
        loss.backward()
        optimizer.step()

        # every 10 batches, print the loss function
        if idx_batch % 10 == 0:
            print("train epoch = " + str(epoch) + " | batch = " + str(idx_batch) + " | loss = "+str(loss.item()))
        train_loss += loss.item()
        count += 1
    train_loss /=count

    # validation loop
    unet.eval()
    val_loss = 0
    count = 0
    for idx_batch, (imagergb, labelmask, filename) in enumerate(val_loader):
        with torch.no_grad(): # no gradient computation required during validation loop
            x = imagergb.to(device)
            y_ = labelmask.to(device)
            y = unet(x)
            loss = loss_func(y, y_)
            val_loss += loss.item()
            count += 1

            # every 5 batches, we also saved the validation label map
            if idx_batch % 5 == 0:
                for idx in range(0, y.shape[0]):
                    max_index = torch.argmax(y[idx], dim=0).cpu().int().numpy()
                    label_rgb_vis = from_label_to_rgb(max_index)
                    gt_correct_format = y_[idx].cpu().int().numpy()
                    gt_vis = from_label_to_rgb(gt_correct_format)
                    result_vis = np.concatenate((label_rgb_vis, gt_vis), axis=1)
                    result_vis_for_saving = (result_vis * 255.0).astype(np.uint8)
                    imsave(train_result_dir + "valdation_seg_result_{}_{}_{}.png".format(epoch, idx_batch, idx), result_vis_for_saving)
    val_loss /=count
    print('epoch ' + str(epoch) + ', training loss = ' + str(train_loss) + ', val_loss = ' + str(val_loss))
    model_location = train_result_dir + "model_file_epoch_" + str(epoch) + "_val_loss="+ str(val_loss) + ".pt"
    torch.save(unet, model_location)
```
